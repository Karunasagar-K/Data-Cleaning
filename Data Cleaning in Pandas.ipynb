## Data Cleaning

### To Check for Duplicates

```python
a[a.duplicated()]
```

### To Drop the Duplicates

```python
a=a.drop_duplicates()
a
```

### To Drop Unwanted Columns

```python
a = a.drop(columns='Not_Useful_Column')
a
```

```python
#To remove the _/.
a['Last_Name'] = a['Last_Name'].str.strip('/._')
a['Last_Name']
```

```python
a['Phone_Number']
```

```python
a['Phone_Number'] = a['Phone_Number'].str.replace('[-/|]','')
a['Phone_Number']
```

```python
a['Phone_Number'] = a['Phone_Number'].str.replace(r'(\d{3})(\d{3})(\d{4})',r'\1-\2-\3')
a['Phone_Number'] 
```

```python
a['Phone_Number'] = a['Phone_Number'].str.replace('NaN','')
a['Phone_Number']
```

```python
a
```

```python
a[['Street_Address','City','Zip_code']] = a['Address'].str.split(',',2,expand=True)
a
```

```python
a['Paying Customer'] = a['Paying Customer'].replace({'Yes':'Y','No':'N'})
a['Do_Not_Contact'] = a['Do_Not_Contact'].replace({'Yes':'Y','No':'N'})
```

```python
a
```

```python
a = a.replace({'N/a':'','NaN':''})
a = a.fillna('')
```
