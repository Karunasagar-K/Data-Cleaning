* %Y: Year with century as a decimal number.
* %m: Month as a zero-padded decimal number.
* %d: Day of the month as a zero-padded decimal number.
* %H: Hour (24-hour clock) as a zero-padded decimal number.
* %M: Minute as a zero-padded decimal number.
* %S: Second as a zero-padded decimal number.
* %f: Microsecond as a decimal number, zero-padded on the left.
* %p: Equivalent of either 'AM' or 'PM' based on the given time.
* %A: Full weekday name.
* %B: Full month name.
* %Z: Time zone name.

## To Find Total Duration of Movies and TV Shows

```python
total_duration = 0

for index, row in nf.iterrows():
    if row['type'] == 'Movie':
        total_duration += int(row['duration'].split()[0])
    elif row['type'] == 'TV Show':    
        seasons = int(row['duration'].split()[0])  
        avg_season_duration = 45
        total_duration += seasons * avg_season_duration
        
print("Total duration:", total_duration, "minutes")        
```
